**Q1. What is hypothesis testing in statistics?**

--> Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating two competing hypotheses—the null hypothesis (H₀) and the alternative hypothesis (H₁)—and using statistical tests to determine which is supported by the data. Hypothesis testing helps determine if observed results are due to chance or if there is a statistically significant effect.

**Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

--> The null hypothesis (H₀) is a statement that assumes no effect, no difference, or no relationship in the population being studied. It serves as the baseline or default assumption.
The alternative hypothesis (H₁) contradicts the null hypothesis and proposes that there is an effect, difference, or relationship. For example:

H₀: The new drug has no effect.

H₁: The new drug has an effect.

**Q3. What is the significance level in hypothesis testing, and why is it important?**

--> The significance level (denoted as α) is the threshold used to decide whether to reject the null hypothesis. Commonly set at 0.05, it represents the probability of making a Type 1 error (rejecting H₀ when it is true). A lower α reduces the likelihood of a Type 1 error but increases the chance of a Type 2 error. It is crucial for determining the rigor of the test.

**Q4. What does a P-value represent in hypothesis testing?**

--> The P-value is the probability of obtaining results as extreme as those observed (or more extreme) if the null hypothesis is true. It measures the strength of evidence against H₀. A smaller P-value indicates stronger evidence to reject H₀.


**Q5. How do you interpret the P-value in hypothesis testing?**


--> If P-value ≤ significance level (e.g., 0.05), reject H₀ (results are statistically significant).

If P-value > significance level, fail to reject H₀ (no significant evidence against it).

For example, a P-value of 0.03 means there’s a 3% chance the observed result occurred under H₀.

**Q6. What are Type 1 and Type 2 errors in hypothesis testing?**


--> Type 1 error: Rejecting H₀ when it is true (false positive). Its probability is controlled by the significance level (α).

Type 2 error: Failing to reject H₀ when it is false (false negative). Its probability is denoted as β, and power = 1 - β represents the likelihood of avoiding a Type 2 error.

**Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**


--> One-tailed test: Tests for an effect in one specific direction (e.g., greater than or less than a value).

Two-tailed test: Tests for an effect in either direction (e.g., not equal to a value).
Two-tailed tests are more conservative since they split the significance level across both tails.

**Q8. What is the Z-test, and when is it used in hypothesis testing?**


--> The Z-test is used to test hypotheses about population means or proportions when the population variance is known and the sample size is large (n > 30). It assumes a normal distribution. For example, comparing a sample mean to a known population mean.

**Q10. What is the T-distribution, and when should it be used instead of the normal distribution?**


--> The T-distribution is a probability distribution similar to the normal distribution but has heavier tails. It is used when the sample size is small (n < 30) or when the population variance is unknown. It accounts for additional uncertainty due to smaller sample sizes.

**Q11. What is the difference between a Z-test and a T-test?**


--> Z-test: Used for large samples or known population variance. Assumes normality.

T-test: Used for small samples or unknown population variance. Relies on the T-distribution.

**Q12. What is the T-test, and how is it used in hypothesis testing?**


--> The T-test evaluates whether the means of two groups are significantly different. Common types include:

One-sample T-test: Compares a sample mean to a known value.

Independent T-test: Compares means of two independent groups.

Paired T-test: Compares means of the same group at two time points.

**Q13. What is the relationship between Z-test and T-test in hypothesis testing?**


--> Both compare sample means to determine if there’s a significant difference. The Z-test is a special case of the T-test when the sample size is large, or population variance is known.

**Q14. What is a confidence interval, and how is it used to interpret statistical results?**


--> A confidence interval is a range of values that is likely to contain the population parameter, with a specified level of confidence (e.g., 95%). It provides a measure of estimate precision and is used alongside hypothesis testing.

**Q15. What is the margin of error, and how does it affect the confidence interval?**


-->The margin of error is the maximum expected difference between the sample statistic and the population parameter. It widens or narrows the confidence interval, depending on sample size and variability.

**Q16. How is Bayes' Theorem used in statistics, and what is its significance?**


--> Bayes' Theorem calculates conditional probabilities, updating prior beliefs with new evidence. It is widely used in fields like machine learning and medical diagnosis to combine prior knowledge with observed data.


**Q17. What is the Chi-square distribution, and when is it used?**


--> The Chi-square distribution is used in categorical data analysis, such as testing independence in contingency tables or goodness of fit for distributions. It applies when data are frequencies.

**Q19. What is the F-distribution, and when is it used in hypothesis testing?**


--> The F-distribution is used to compare variances between two datasets. It is applied in ANOVA and regression analysis to test the ratio of variances.

**Q20. What is an ANOVA test, and what are its assumptions?**


--> ANOVA (Analysis of Variance) tests differences between three or more group means. Assumptions:

Groups are independent.

Data are normally distributed.

Variances across groups are equal (homogeneity of variances).

**Q21. What are the different types of ANOVA tests?**


--> One-way ANOVA: Tests one factor with multiple levels.

Two-way ANOVA: Tests two factors and their interaction.

Repeated measures ANOVA: Tests the same subjects across conditions.

**Q22. What is the F-test, and how does it relate to hypothesis testing?**


--> The F-test compares variances to test hypotheses about means (e.g., in ANOVA). It uses the ratio of two variances to determine if they are significantly different.

```python
# 1. Perform a Z-test for comparing a sample mean to a known population mean.

from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size):
    z_score = (sample_mean - population_mean) / (population_std / (sample_size ** 0.5))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    return z_score, p_value

z, p = z_test(105, 100, 15, 30)
print(f"Z-score: {z}, P-value: {p}")
if p < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# 2. Simulate random data and perform hypothesis testing.

import numpy as np
from scipy.stats import ttest_1samp

np.random.seed(42)
data = np.random.normal(100, 10, 50)
t_stat, p_value = ttest_1samp(data, 100)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
# 3.  One-sample Z-test.

z, p = z_test(np.mean(data), 100, 10, len(data))
print(f"Z-score: {z}, P-value: {p}")
```

```python
# 4. Two-tailed Z-test with visualization.

import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label="Standard Normal Curve")
plt.axvline(-1.96, color='red', linestyle='--', label="Critical Value (-1.96, 1.96)")
plt.axvline(1.96, color='red', linestyle='--')
plt.fill_between(x, y, where=(x <= -1.96) | (x >= 1.96), color='red', alpha=0.3, label="Rejection Region")
plt.legend()
plt.title("Two-Tailed Z-test Decision Regions")
plt.show()
```

```python
# 5. Type 1 and Type 2 errors visualization.

def type1_type2_error(alpha, beta, mu0, mu1, sigma):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    null_dist = norm.pdf(x, mu0, sigma)
    alt_dist = norm.pdf(x, mu1, sigma)

    plt.plot(x, null_dist, label="Null Distribution")
    plt.plot(x, alt_dist, label="Alternative Distribution")
    plt.axvline(mu0 + norm.ppf(1 - alpha) * sigma, color='red', linestyle='--', label="Critical Value")
    plt.fill_between(x, null_dist, where=(x > mu0 + norm.ppf(1 - alpha) * sigma), color='red', alpha=0.3, label="Type 1 Error")
    plt.fill_between(x, alt_dist, where=(x <= mu0 + norm.ppf(1 - alpha) * sigma), color='blue', alpha=0.3, label="Type 2 Error")
    plt.legend()
    plt.show()

type1_type2_error(0.05, 0.2, 100, 110, 15)
```

```python
# 9. Confidence Interval Calculation.

def confidence_interval(sample_mean, sample_std, sample_size, confidence_level=0.95):
    z = norm.ppf((1 + confidence_level) / 2)
    margin_of_error = z * (sample_std / (sample_size ** 0.5))
    lower = sample_mean - margin_of_error
    upper = sample_mean + margin_of_error
    return lower, upper

ci = confidence_interval(105, 10, 30)
print(f"Confidence Interval: {ci}")
```

```python
# 10. Margin of Error Calculation.

def margin_of_error(sample_std, sample_size, confidence_level=0.95):
    z = norm.ppf((1 + confidence_level) / 2)
    return z * (sample_std / (sample_size ** 0.5))

moe = margin_of_error(10, 30)
print(f"Margin of Error: {moe}")
```

```python
# 11. Bayesian Inference.

def bayesian_inference(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.5
likelihood = 0.7
evidence = 0.6
posterior = bayesian_inference(prior, likelihood, evidence)
print(f"Posterior Probability: {posterior}")
```

```python
# 16. Implement an F-test for variance comparison.

import numpy as np

def f_test(group1, group2):
    var1 = np.var(group1, ddof=1)
    var2 = np.var(group2, ddof=1)
    f_stat = var1 / var2
    return f_stat

group1 = [10, 12, 13, 15, 17]
group2 = [8, 9, 11, 10, 14]
f_stat = f_test(group1, group2)
print(f"F-statistic: {f_stat}")
```

```python
# 19. Check ANOVA assumptions (normality, independence, and equal variance).

from scipy.stats import shapiro, levene

group1 = [85, 90, 88, 86, 89]
group2 = [92, 95, 91, 93, 94]
group3 = [78, 75, 80, 76, 79]

print("Shapiro-Wilk Test for Normality:")
for i, group in enumerate([group1, group2, group3], 1):
    stat, p = shapiro(group)
    print(f"Group {i}: P-value = {p}")

stat, p = levene(group1, group2, group3)
print(f"Levene’s Test: P-value = {p}")
```

```python
# 20. Two-way ANOVA test with visualization.

import statsmodels.api as sm
from statsmodels.formula.api import ols

data = {
    "Factor1": ["A", "A", "A", "B", "B", "B", "C", "C", "C"] * 2,
    "Factor2": ["X", "X", "X", "Y", "Y", "Y", "Z", "Z", "Z"] * 2,
    "Value": [85, 88, 90, 75, 78, 80, 95, 98, 100, 90, 92, 95, 85, 87, 88, 100, 102, 104]
}
df = pd.DataFrame(data)

# Two-way ANOVA
model = ols('Value ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

# Visualization
sns.boxplot(x="Factor1", y="Value", hue="Factor2", data=df)
plt.title("Two-Way ANOVA: Interaction Plot")
plt.show()
```

```python
# 24. Perform a hypothesis test for population variance using Chi-square distribution.

from scipy.stats import chi2

sample = [10, 12, 15, 14, 13, 11, 10, 14, 13, 12]
sample_size = len(sample)
sample_variance = np.var(sample, ddof=1)
population_variance = 9

chi_stat = (sample_size - 1) * sample_variance / population_variance
p_value = 1 - chi2.cdf(chi_stat, df=sample_size - 1)

print(f"Chi-square Statistic: {chi_stat}, P-value: {p_value}")
```

```python
# 25. Implement an F-test for comparing the variances of two datasets.

import numpy as np
from scipy.stats import f

group1 = [8, 10, 12, 14, 10, 13, 11]
group2 = [15, 18, 16, 20, 19, 17]

var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)

f_stat = var1 / var2
df1 = len(group1) - 1
df2 = len(group2) - 1

p_value = 1 - f.cdf(f_stat, df1, df2)

print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

```python
# 27. Perform a one-way ANOVA test in Python with boxplots to compare group means.

from scipy.stats import f_oneway
import seaborn as sns
import pandas as pd

data = {
    "Group": ["A"] * 10 + ["B"] * 10 + ["C"] * 10,
    "Values": np.random.normal(loc=50, scale=5, size=10).tolist() +
              np.random.normal(loc=55, scale=5, size=10).tolist() +
              np.random.normal(loc=60, scale=5, size=10).tolist()
}
df = pd.DataFrame(data)

groups = [df[df["Group"] == g]["Values"] for g in df["Group"].unique()]
f_stat, p_value = f_oneway(*groups)
print(f"F-statistic: {f_stat}, P-value: {p_value}")

sns.boxplot(x="Group", y="Values", data=df)
plt.title("One-Way ANOVA: Group Means Comparison")
plt.show()
```
